```python
f = open("TestingFile.txt", "r+")
content = f.read()      # read first
f.write("\nHello")        # overwrites from current cursor
f.close()
```

```python
f = open("data.txt", "w+")
f.write("\nNew data")     # old data gone
f.seek(0)               # move cursor to start to read
print(f.read())
f.close()
```

```python
f = open("data.txt", "a+")
f.write("Added line\n")  # always writes at end
f.seek(0)                # must move cursor to read
print(f.read())
f.close()
```
