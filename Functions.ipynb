#### Type of Arguments
- Positional Arguments
- Default Arguments
- Keyword Arguments
- Variable Length Arguments
- Variable Keyword Arguments

```python
# 1. Positional Arguments
# function defination
def division(a,b):
    if b==0:
        return "Zero division not possiable"
    return a / b
#Function calling
c,d = 10,5
res = division(d,c)
print(res)
```

```python
# 2. Keyword Arguments
# function defination
def division(a,b):
    if b==0:
        return "Zero division not possiable"
    return a / b
#Function calling
c,d = 10,5
res = division(b = d,a = c)
print(res)
```

```python
## 3. Default arguments
#function defination
def addition(a,b,c=0,d=0):
    return a+b+c + d
# function calling
res1 = addition(10,15)
res2 = addition(10,15,-5)
res3 = addition(10,15,10,10)
print(res1, res2, res3)
```

```python
# variable length arguments
# function defination
def addition(*args):
    print(type(args))
    print(args)
    s = 0
    for num in args:
        s = s + num
    return s

# Function calling
res = addition(10,15,10,20,30,40,1,2,3,5)
print(res)
```

```python
# variable length keyword arguments
def addition(a,**kwargs):
    print(type(kwargs))
    print(a, kwargs)
    s = 0
    for val in kwargs.values():
        s += val
    return s
print(addition(40,b = 20,c = 20,e =30))
```

##### Scope of Variables
- Local Variable -The variable can access the inside function, then those variable called as `Local Variables`
- Global Variable - The variables are defined out side of the function and we can access those values inside the function but can't modify
- Non Local variable -The variable can defined inside nested function , those variable calld as `Non local variabl`

```python
def fun():
    x = 5 # Local variable
    x = x+5
    print(x)
fun()
print(x)
```

```python
x = 10 
def fun():
    print(x)
fun()    
x = x + 1
print(x)
```

```python
# Global Variable - The variable are defined out side the function and 
# we can access those values inside the function but can't modify
x = 10 # Gobal variable 
def fun():
    x = x+2  # We cant modify the gobal variables inside function  
    print(x) # we can access the gobal variables inside the function 
fun()  
x = x + 1
print(x)
```

```python
# Global Variable modify inside function by using 'global' keyword
x = 10 # Gobal variable 
def fun():
    global x
    x = x+2  
    print(x) 
fun()  
x = x + 1
print(x)
```

```python
# Non Local variable -The variable can defined inside nested function ,
#those variable calld as `Non local variabl`
x = "Global" # Gobal variable
def outer():
    y = "Outer" # local variable
    def inner():
        z = "Inner" # Non Local variable
        print(z, y, x)
    inner()
    print(y,x)
outer()
print(x)
```

```python
def first_n_even(n):
    l =[]
    for i in range(n):
        l.append(2*i)
    return l
res = first_n_even(10)
print(res)
```

```python
def first_n_odd(n):
    l =[]
    s = 0
    for i in range(n):
        l.append(2*i+1)
        s += 2*i+1
    return l, s
    
l, s = first_n_odd(10)
print(s)
print(l)
print(*l)
```
