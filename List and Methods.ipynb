## List Methods

- append(value) - it add the single value at end of list
- extend(iterable) - it add the iterable values at end of list
- insert(position, value) - add the value at specified position and followed values are right shifted
- index(vlaue) - returns the first position of value, otherwise `ValueError`
- remove(value) - it removes the value at first occurance position
- pop() - returns and removes the pop value as per specified position, otherwise pops the last value
- clear() - removes the all elements from list
- del - removes the list
- count(value) - returns the total occurance of a value in the list
- len(list_name) - total elements in the list or length of list
- min(list_name) - returns the min value from the list
- max(list_name) - returns the max value from the list
- sum(list_name) - sums the all elements in the list


```python
Lst = [10,34,45,21,56,3,2,4]
mini = float('inf')
for element in Lst:
    if mini>element:
        mini = element
print(mini)
```

```python
Lst = [10,34,45,21,56,3,2,4]
maxi = float('-inf')
for element in Lst:
    if maxi<element:
        maxi = element
print(maxi)
```

```python
Lst = [10,34,45,21,56,3,2,4]
total = 0
for element in Lst:
    total += element
print(total)
```

```python
## append()
l = [1, 2,3,5.5,'a','b', 1,2,3]
print(l)
l.append(100)
print("After append Operation:", l)
```

```python
## extend()
l = [1, 2,3,5.5,'a','b', 1,2,3]
l.extend([200,300])
print("After extend Operation:", l)
```

```python
## insert()
l = [1, 2,3,5.5,'a','b', 1,2,3]
l.insert(4,'Codegnan')
print("After insert Operation:", l)
```

```python
# remove(), pop(3), clear(), del
l = [1, 2, 3, 5.5, 'Codegnan', 'a', 'b', 1, 2, 3]
print("Original list:", l)
l.remove(3)
print("After remove operation:", l)
l.pop()
print("After pop operation:", l)
l.pop(3)
print("After pop with position:", l)
del l[0]
print(l)
del l[2:5]
print("After del operation:",l)
l.clear()
print("After clear opration:",l)
```

```python
# min(), max(), sum()
l = [1,2, 29,43, 3, 10, 200]
min_val = min(l)
max_val = max(l)
sum_list = sum(l)
print("Min value in the list:", min_val)
print("max value in the list:", max_val)
print("sum of list:", sum_list)
```

```python
#reverse() - it reverse the original list in reverse order
l = [1,2, 29,43, 3, 10, 200]
l.reverse()
print("Reverse of list:", l)
```

```python
# sort() - it sorts the original list in particular order
# assending order sort
l = [1,2, 29,43, 3, 10, 200]
l.sort()
print(l)
```

```python
# descending order sort
l = [1,2, 29,43, 3, 10, 200]
l.sort(reverse = True)
print(l)
```

```python
# sorted() -it sort the list in perticular order and it returns new sorted list of values
l = [1,2, 29,43, 3, 10, 200]
new_list = sorted(l)
print(new_list)
```

```python
## count(value), index(value)

l = [1, 2, 3, 10, 29, 43, 200,23,2,3,2]
count_2 = l.count(2)
print("Total 2 element count:", count_2)
ind = l.index(200)
print("29 value index:", ind)
```
