## Strings

- it is sequence of characters
- it is denote with `''` or `""`
- Strings are immutable
- strings are index based and Sliceable
  

```python
s = 'abcde'
# indexing
print(s[0])
print(s[1])
print(s[-1])
```

```python
# Slicing
s = 'abcde'
print(s[1:4])
print(s[2:3])
print(s[:3])
print(s[:])
print(s[:-2])
print(s[-3:])
print(s[-4:-1])
print(s[::2])
```

```python
# string reversal using slicing
print(s[::-1])
```

```python
## String Concatination by using ' + ' 
s1 = "Python"
s2 = "Programming"
print(s1 +" " + s2)
```

```python
## String Concatination by using ' + ' 
s1 = "Wellcome"
s2 = "to"
s3 = "Programming"
print(s1 + " " + s2 +" " + s3)
```

```python
## String Case Conversion 
string = "Nandini"
s2 = "i am a Python TraineR"
# lower() - it converts string into lower case 
print("Lower case coversion:", string.lower())
# Upper() - It converts the string into Uppder case
print(string.upper())
# swapcase() - it coverts upper to lower and lower to upper
print(string.swapcase())
# Title() - it converts the every first letter into upper case
print(s2.title())
# capitalize() - It coverts the every first letter in string into upper
print(s2.capitalize())
```

## checking Functions
- isalpha() -returns True if string contains all are alphabet values
- isalnum() - return True if string contains alphabets or digits
- isdigit() - return True if string contains all are digit values

```python
s1 = 'abcd'
s2 = 'abc123'
s3 = '1234'
s4 = 'abc@12'
print(s1.isalpha(), s1.isalnum(), s1.isdigit())
print(s2.isalpha(), s2.isalnum(), s2.isdigit())
print(s3.isalpha(), s3.isalnum(), s3.isdigit())
print(s4.isalnum())
```
