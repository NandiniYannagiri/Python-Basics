# Constructor

# Types of Methods in a Class

##### Instance Method → works on object (needs self).

##### Class Method → works on class (uses @classmethod and cls).

##### Static Method → utility function (uses @staticmethod, no self or cls)

```python
class Student:
    school = "ABC School"   # class variable
    
    def __init__(self, name, marks):   # instance method
        self.name = name
        self.marks = marks
    
    def show(self):   # instance method
        print(f"{self.name} got {self.marks} marks")
    
    @classmethod
    def school_info(cls):   # class method
        print("School Name:", cls.school)
    
    @staticmethod
    def greet():   # static method
        print("Welcome to the Student Portal!")
        

s1 = Student("Nandu", 90)
s1.show()              # Nandu got 90
Student.school_info()  # School Name: ABC School
Student.greet()        # Welcome to the Student Portal!
```

### Another example

```python
class Student:
    # Class variable (shared by all objects)
    school = "ABC Public School"
    
    # Constructor (__init__) → runs automatically on object creation
    def __init__(my_own, name, marks):
        my_own.name = name      # Instance variable
        my_own.marks = marks    # Instance variable
    
    # Instance method → works with object data
    def show_details(my_own):
        print(f"Name: {my_own.name}, Marks: {my_own.marks}")
    
    # Class method → works with class-level data
    @classmethod
    def get_school(c):
        print("School Name:", c.school)
    
    # Static method → independent utility, doesn’t use self or cls
    @staticmethod
    def is_pass(m):
        return m >= 40


# -----------------------
# Creating objects
s1 = Student("Nandu", 85)
s2 = Student("Ravi", 35)

# Using instance method
s1.show_details()   # Name: Nandu, Marks: 85
s2.show_details()   # Name: Ravi, Marks: 35

# Using class method
Student.get_school()   # School Name: ABC Public School

# Using static method
print(Student.is_pass(90))  # True
print(Student.is_pass(25))  # False
```

# Important points to note

#####  A class can have multiple objects.

#####  __init__ runs automatically when you create an object.

#####  Objects can access both instance variables and class variables.

#####  Methods define behaviors of objects.

```python
class ATM:
    bank_name = "HDFC Bank"   # class variable
    
    def __init__(self, account_no, balance):
        self.account_no = account_no
        self.balance = balance
    
    def deposit(self, amount):   # instance method
        self.balance += amount
        print(f"Deposited {amount}, New Balance = {self.balance}")
    
    def withdraw(self, amount):  # instance method
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawn {amount}, Remaining Balance = {self.balance}")
        else:
            print("Insufficient funds")
    
    @classmethod
    def bank_info(cls):   # class method
        print("Bank Name:", cls.bank_name)


# Creating objects
cust1 = ATM(101, 5000)
cust2 = ATM(102, 10000)
cust1.

cust1.deposit(2000)   # Deposited 2000, New Balance = 7000
cust1.withdraw(3000)  # Withdrawn 3000, Remaining Balance = 4000

cust2.deposit(5000)
cust2.withdraw(3000)

ATM.bank_info()       # Bank Name: HDFC Bank
cust1.
```

# Encapsulation

##### Python does not have strict access modifiers (private, public) like Java, but it follows naming conventions:

##### Public → accessible everywhere.

##### Protected (_var) → intended for subclass use (still accessible outside, but discouraged).

##### Private (__var) → name-mangled, harder to access directly.

```python
class Student:
    def __init__(self, name, marks):
        self._name = name         # protected variable
        self.__marks = marks      # private variable

    # Getter for private
    def get_marks(self):
        return self.__marks

    # Setter for private
    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print("Invalid marks!")

    def display(self):
        print("Name:", self._name, "Marks:", self.__marks)


# Using class
s1 = Student("Nandu", 85)
s1.

# Access protected variable (possible, but not recommended)
print("Accessing protected:", s1._name)

# Access private variable directly ( not allowed)
# print(s1.__marks)   # AttributeError

# Proper way: use getter/setter
print("Private via getter:", s1.get_marks())
s1.set_marks(95)
s1.display()
```

# Inheritance

##### Inheritance is an OOP concept where one class (child/derived) can use the properties and methods of another class (parent/base).
##### This helps in code reuse and makes programs easier to maintain.

```python
class Parent:
    def show_parent(self):
        print("I am Parent")

class Child(Parent):   # Child inherits from Parent
    def show_child(self):
        print("I am Child")

c = Child()
c.show_parent()  # Parent method
c.show_child()   # Child method
```

# Types of Inheritance in Python

##### 1. Single Inheritance
##### 2. Multiple Inheritance
##### 3. Multi level Inheritance
##### 4. Hierarchical Inheritance
##### 5. Hybrid Inheritance

```python
# 1.Single Inheritance - A child inherits from one parent.

class Vehicle:   # Parent
    def drive(self):
        print("This vehicle can be driven")

class Car(Vehicle):   # Child
    def fuel(self):
        print("This car uses petrol/diesel")

c = Car()
c.drive()  # from Vehicle
c.fuel()   # from Car
```

```python
# 2. Multiple inheritance - A child inherits from multiple parents.

class Engine:
    def engine_info(self):
        print("Engine: 1500cc")

class Tyres:
    def tyres_info(self):
        print("Tyres: Alloy wheels")

class Car(Engine, Tyres):  # Child inherits both
    def show(self):
        print("This is a Car")

c = Car()
c.engine_info()
c.tyres_info()
c.show()
```

```python
# 3. Multilevel Inheritance - Inheritance across multiple levels (grandparent → parent → child).

class LivingBeing:   # Grandparent
    def breathe(self):
        print("Breathing...")

class Animal(LivingBeing):   # Parent
    def eat(self):
        print("Eating...")

class Dog(Animal):   # Child
    def bark(self):
        print("Barking...")

d = Dog()
d.breathe()
d.eat()
d.bark()
```

```python
# 4. Hierarchical Inheritance - Multiple children inherit from the same parent.

class Shape:   # Parent
    def area(self):
        print("Calculating area...")

class Circle(Shape):   # Child 1
    def circle_area(self, r):
        print("Circle area:", 3.14 * r * r)

class Rectangle(Shape):   # Child 2
    def rectangle_area(self, l, w):
        print("Rectangle area:", l * w)

c = Circle()
r = Rectangle()

c.area()
c.circle_area(5)

r.area()
r.rectangle_area(4, 6)
```

```python
# 5. Hybrid Inheritance - Combination of different inheritance types.

class Person:
    def details(self):
        print("I am a Person")

class Employee(Person):   # single inheritance
    def emp_details(self):
        print("I am an Employee")
        
class Student(Person):     # hierarchal inheritance  
    def stu_details(self): 
        print("I am a Student")

class WorkingStudent(Employee, Student):  # Multiple Inheritance
    def ws_details(self):
        print("I am a Working Student")

ws = WorkingStudent()
ws.details()       # From Person
ws.emp_details()   # From Employee
ws.stu_details()   # From Student
ws.ws_details()    # Own method
```

# Polymorphism

##### Polymorphism means “many forms”.
##### In OOP, it allows the same method name to behave differently depending on the object or context.
##### It increases flexibility and code reusability.

#### Types of Polymorphism in Python

```python
 # 1. Polymorphism with Functions/Operators (Built-in) -Some operators/functions work differently based on data types.

print(5 + 10)       # integer addition → 15
print("Hi" + "All") # string concatenation → HiAll
print([1,2] + [3])  # list merge → [1, 2, 3]

```

```python
# 2. Polymorphism with Functions (Duck Typing) - Python follows duck typing: if an object has a method, it can be used, regardless of its class.
#Both classes have sound(), but the behavior differs.

class Dog:
    def sound(self):
        return "Bark"

class Cat:
    def sound(self):
        return "Meow"

def animal_sound(animal):
    print(animal.sound())

dog = Dog()
cat = Cat()

animal_sound(dog)  # Bark
animal_sound(cat)  # Meow
```

```python
# 3. Polymorphism with Abstract Classes (Runtime Polymorphism) - Force subclasses to implement methods → achieved via abc module.
#Same method fly(), but behavior depends on the object type.

from abc import ABC, abstractmethod

class Bird(ABC):
    @abstractmethod
    def fly(self):
        pass

class Sparrow(Bird):
    def fly(self):
        print("Sparrow can fly high")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

birds = [Sparrow(), Penguin()]
for b in birds:
    b.fly()
```

```python
# 4. Polymorphism with Method Overloading (Not Native in Python)
#Python does not support traditional method overloading, but we can simulate it with default arguments or *args.
class Math:
    def add(self, a=0, b=0, c=0):
        return a + b + c

m = Math()
print(m.add(2))
print(m.add(2, 3))
print(m.add(2, 3, 4)) 
```

```python
# 5. Polymorphism with Inheritance (Method Overriding) - Child class can override parent’s method.
# Same method name, different implementation.

class Vehicle:
    def fuel_type(self):
        return "Diesel or Petrol"

class ElectricCar(Vehicle):
    def fuel_type(self):    # Overriding
        return "Electricity"

v = Vehicle() 
e = ElectricCar()

print(v.fuel_type())  # Diesel or Petrol
print(e.fuel_type())  # Electricity
```

```python
# another example for method overriding

class Vehicle:
    def __init__(self, brand):
        self.brand = brand
        print(f"Vehicle created: {brand}")

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)   # Call parent constructor
        self.model = model
        print(f"Car created: {brand}, {model}")

c = Car("Tesla", "Model S")
```

## Abstraction

##### Abstraction is hiding the implementation details and showing only the essential features.
##### It focuses on what an object does, not how it does it.

##### In Python, abstraction is implemented using:

##### Abstract Classes (abc module)

##### Interfaces (Python doesn’t have a keyword, but abstract classes act like interfaces)

##### Why Abstraction?

##### Provides security (hides background details).

##### Simplifies complex systems.

##### Promotes reusability.

##### Enforces implementation rules for subclasses.

```python
from abc import ABC, abstractmethod

class Shape(ABC):  # abstract class
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass


class Circle(Shape):  # child class
    def __init__(self, r):
        self.r = r
    
    def area(self):  # must implement
        return 3.14 * self.r * self.r
    
    def perimeter(self):  # must implement
        return 2 * 3.14 * self.r
class Square(Shape):
    def __init__(self, l):
        self.l = l
    def area(self):
        return self.l*self.l
    def perimeter(self):
        return 4*self.l

class Rectangle(Shape):
    def __init__(self, l,b):
        self.l = l
        self.b = b
    def area(self):
        return self.l*self.b
    def perimeter(self):
        return 2*(self.l+self.b)
c = Circle(5)
print(c.area())       # 78.5
print(c.perimeter())  # 31.4

s = Square(4)
print(s.area())
print(s.perimeter())

s = Rectangle(4,3)
print(s.area())
print(s.perimeter())
```
