## Dictionary
- Dictionary is ordered collection of key value pairs
- `{key : value}`
- In dictionary keys are unique and values allows deplicates
- Dictionary is defiuned with `{}`
- It is non index based
- Values are accessed with keys

```python
d = { 1: 'a', 'A':2, 6: 4, (1,2): 4, 5:[1,2,(3,4),5]}
```

```python
# empty dictionary
d = {}
d1 = dict()
# dictionary with vales 
d3 = { 1: 'a', 'A':2, 6: 4, (1,2): 4, 5:[1,2,(3,4),5]}
print(type(d), type(d1), type(d3))
```

```python
# dictionary values access by using keys
d = { 1: 'a', 'A':2, 6: 4, (1,2): 4, 5:[1,2,(3,4),5]}
print(d[1])
print(d['A'])
print(d[5])
```

## Dictionary Methods

- keys() - Returns the keys of dictionary
- values() - Returns values of dictionary
- items() - returns list of tuple of key values pairs

```python
d = {'language': 'python', 'version': '3.13.7', 'level':"High level"}
keys = d.keys()
values = d.values()
items = d.items()
print(type(keys), type(values), type(items))
print(keys)
print(values)
print(items)
```

```python
d = {'language': 'python', 'version': '3.13.5', 'level':"High level"}
keys = d.keys()

print(type(keys))
print(keys)
list_keys = list(keys)
print(type(list_keys))
print(list_keys)
print(list_keys[1])
```

```python
d = {'language': 'python', 'version': '3.13.5', 'level':"High level"}

items = d.items()
print(type(items))
print(items)
# dict_items to list type conversion
items_list = list(items)
print(type(items_list))
print(items_list)
a, b = items_list[1]
print(a, b)
```

- pop(key) - return and removes the value of key if key is present in the dictionary, otherwise error
- popitem() - return and removes the last updated item from dictionary

```python
## pop(key)
d = {'language': 'python', 'version': '3.13.5', 'level':"High level"}
val = d.pop('version')
print(val)
print(d)
```

```python
## popitem()
d = {'language': 'python', 'version': '3.13.5', 'level':"High level"}
item = d.popitem()
print(item)
print(d)
```

```python
## Accessing and updating the values
d = {'language': 'python', 'version': '3.13.5', 'level':"High level"}
d['A'] = 1
d['language'] = 'Java'
print(d)
```

- get(key, default) - Returns the value of key, otherwise default value
- update(dictionary) - It updates dictionary with new items 

```python
# get(key, default)
d = {'language': 'python', 'version': '3.13.5', 'level':"High level"}
val = d.get('updted_version', 'key is not in dictionary')
print(val)
```

```python
# get(key, default)
d = {'language': 'python', 'version': '3.13.5', 'level':"High level"}
val = d.get('Z',10)
print(val)
```

```python
# update(dictionary)
d1 = {'language': 'python', 'version': '3.13.5'}
d2 = {'language': 'java', 'level':"High level"}
d1.update(d2)
print(d1)
print(d2)
```

```python
# update(dictionary)
d1 = {'language': 'python', 'version': '3.13.5'}
d2 = {'language': 'java', 'level':"High level"}
d2.update(d1)
print(d1)
print(d2)
```

```python
# clear() - Removes the all items from dictionary
d1 = {'language': 'python', 'version': '3.13.5'}
d1.clear()
print(d1)
```
